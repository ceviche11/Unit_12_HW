## 1. Sentiment Analysis

Use the [newsapi](https://newsapi.org/) to pull the latest news articles for Bitcoin and Ethereum and create a DataFrame of sentiment scores for each coin.

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
import nltk as nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from newsapi import NewsApiClient
analyzer = SentimentIntensityAnalyzer()
from nltk.util import ngrams


%matplotlib inline
```

```python
# Read your api key environment variable
load_dotenv()
api_key = os.getenv('NEWS_API_KEY') 
type(api_key)
```

```python
# Create a newsapi client

newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch the Bitcoin news articles
bitcoin_articles = newsapi.get_everything(q="Bitcoin", 
                                          language="en",
                                          page_size=100,
                                         sort_by='relevancy')
bitcoin_articles['totalResults']
```

```python
# Fetch the Ethereum news articles
ethereum_articles = newsapi.get_everything(q="Ethereum", 
                                           language="en",
                                           page_size=100,
                                           sort_by='relevancy')
ethereum_articles['totalResults']
```

```python
# Create the Bitcoin sentiment scores DataFrame
bitcoin_sentiments = []

for article in bitcoin_articles["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        bitcoin_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
    
# Stuff into DataFrame 
bitcoin_df = pd.DataFrame(bitcoin_sentiments)
```

```python
# Create the Ethereum sentiment scores DataFrame
ethereum_sentiments = []

for article in ethereum_articles["articles"]:
    try:
        text = article["content"]
        date = article["publishedAt"][:10]
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        
        ethereum_sentiments.append({
            "text": text,
            "date": date,
            "compound": compound,
            "positive": pos,
            "negative": neg,
            "neutral": neu
            
        })
        
    except AttributeError:
        pass
    
ethereum_df = pd.DataFrame(ethereum_sentiments)
```

```python
# Describe the Bitcoin Sentiment
bitcoin_df.describe()
```

```python
# Describe the Ethereum Sentiment
ethereum_df.describe()
```

### Questions:

Q: Which coin had the highest mean positive score?

A: Ethereum had a higher mean positive score 

Q: Which coin had the highest mean compound score?

A: Bitcoin had the highest mean compound score 

Q. Which coin had the highest max positive score?

A: Ethereum had the highest max positive score

###   Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word.
2. Remove Punctuation.
3. Remove Stopwords.

```python
from nltk.tokenize import word_tokenize, sent_tokenize, RegexpTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
bitcoin_df.head()
```

```python
# Instantiate the lemmatizer
lemmatizer = WordNetLemmatizer()

# Create a list of stopwords
print(stopwords.words('english'))

# Expand the default stopwords list if necessary
# Will go back here if necessary when if a stop word is not captured here 
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""

    sw = set(stopwords.words('english'))
    
    #regex = re.compile("[^a-zA-Z ]")
    
    tokenizer = RegexpTokenizer(r'\w+')
    words = tokenizer.tokenize(text)

    # Remove the punctuation from text

    ##re_clean = regex.sub('', text)
    # Create a tokenized list of the words
    
    ##words = word_tokenize(re_clean)
    # Lemmatize words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
    
    # Convert the words to lowercase
    # Remove the stop words
    tokens = [word.lower() for word in lem if word.lower() not in sw]
    
    return tokens

bitcoin_df

# Apply the function to text column 
# Create a new tokens column for Bitcoin
bitcoin_df['tokenized'] = bitcoin_df['text'].apply(tokenizer)

# Display new tokens column  
bitcoin_df['tokenized']
```

```python
# Create a new tokens column for Ethereum
ethereum_df['tokenized'] = ethereum_df['text'].apply(tokenizer)
```

```python
# Function token_count generates the top 10 words for a given coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Use token_count to get the top 10 words for Bitcoin

# Make a list out of the bitcoin tokens column in the bitcoins_df dataframe 
bitcoin_tokens = bitcoin_df['tokenized'].to_list()

# List is a nested list, so flatten this list to use token_count 
merged_tokens = []

for token in bitcoin_tokens: 
    merged_tokens.append(token)

# Make nested list into single list 
flattened_tokens = []
for sublist in merged_tokens:
    for val in sublist:
        flattened_tokens.append(val)


token_count(flattened_tokens, N=10)
```

```python
# Use token_count to get the top 10 words for Ethereum
ethereum_tokens = ethereum_df['tokenized'].to_list()

merged_tokens_2 = []

for token in ethereum_tokens: 
    merged_tokens_2.append(token)
    
    

flattened_tokens_2 = []
for sublist in merged_tokens_2:
    for val in sublist:
        flattened_tokens_2.append(val)


token_count(flattened_tokens_2, N=10)


```

## 3. Named Entity Recognition

In this section, you will build a named entity recognition model for both Bitcoin and Ethereum, then visualize the tags using SpaCy.

```python


import spacy
from spacy import displacy
```

```python
# Download the language model for SpaCy
!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

---
### Bitcoin NER

```python
# Concatenate all of the Bitcoin text together
# I wanted to select out individual articles, as those led to the best NER visualizations 

concat_btc_text = bitcoin_df['text'].iloc[0]
```

```python
# Run the NER processor on all of the text and add a title 
bitcoin_ner = nlp(concat_btc_text)


```

```python
# Render the visualization
displacy.render(bitcoin_ner, style='ent')
```

```python
# List all Entities
print([ent.text for ent in bitcoin_ner.ents])
```
